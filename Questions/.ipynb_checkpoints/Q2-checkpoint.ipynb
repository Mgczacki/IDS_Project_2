# Question 2:

Are movies that are newer rated differently than movies that are older? [Hint: Do a median split of year of release to contrast movies in terms of whether they are old or new]

Step 1: Finding newer movies:
- Movies that are above the median release year are deemed new, others are deemed old.

```python
def parse_year(s):
    return int(s.split("(")[-1].strip()[0:4])
```

```python
movie_release_years = [parse_year(col) for col in movie_cols]
```

```python
median_release_year = np.median(movie_release_years)
```

```python
new_mask = movie_release_years > median_release_year
```

```python
new_movies = movie_cols[new_mask]
```

```python
new_movies
```

```python
old_movies = [m for m in movie_cols if m not in new_movies]
```

```python
old_movies
```

Step 3: Decide the test design.

- For this use-case, we will assume that sample medians can correctly describe the ratings for movies. Therefore each movie will be assigned the sample median of all its ratings.
- Population parameters are unknown.
- The groups are correlated. People don't only evaluate one movie, and the same people can evaluate movies in both the new and old categories. Therefore it is a within-subjects-design. However, since we will be reducing all ratings for each movie using a median, we will assume that the scores for each movie are mostly independent.

```python
df_new = df[new_movies]
new_scores = df_new.median(axis=0)
```

```python
df_old = df[old_movies]
old_scores = df_old.median(axis=0)
```

Because of our assumptions and the analysis of how ratings should be analyzed, we will be using a Mann-Whitney U Test.

```python
import scipy.stats as stats
```

```python
stats.mannwhitneyu(new_scores, old_scores) 
```

The test yields a resulting p-value of 0.19865156776112602. This gives us a very low degree of confidence that new movies are rated differently, than old ones. Because this is not below our desired confidence value ($\alpha = 0.005$), we have no evidence to reject the null-hypothesis of the medians of both groups being the same.
