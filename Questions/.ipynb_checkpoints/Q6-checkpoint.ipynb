```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.stats.weightstats import ttest_ind
import pingouin as pg
```

```python
data_df = pd.read_csv('../movieReplicationSet.csv')
data_df
```


# Question6: What proportion of movies exhibit an “only child effect”, i.e. are rated different by viewers with siblings vs. those without?


First step: We seperate two groups for the only child and the child with siblings.

```python
only_child_ratings = data_df[data_df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'] == 1].iloc[:,0:400]
only_child_ratings
```

```python
# we reset the index
only_child_ratings.reset_index(inplace = True)
```

```python
only_child_ratings
```

```python
siblings_ratings = data_df[data_df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'] == 0].iloc[:,0:400]
siblings_ratings
```

```python
siblings_ratings.reset_index(inplace = True)
```

```python
siblings_ratings
```

```python
# The rows contain 0-176 are the only child rateings, and 177-1070 are the child with siblings rateings
df_append = only_child_ratings._append(siblings_ratings, ignore_index=True)
df_append
```

```python
df_append = df_append.drop('index', axis=1)
```

```python
df_append
```

```python
#iterate each movies from the newly appended dataset(The rows contain 0-176 are the only child rateings, 
#and 177-1070 are the child with siblings rateings)
#we compare the two groups with Mann-Whitney U Test for each movie iteration, and if the p-value for that movie is smaller than the alpha value(0.005), 
#then the count value will plus one
#and finally, we will use the counted value/sum of the total movies to calcuate the proportion which shows the only child effect. 
count = 0
for column in df_append:
     
    # Select column contents by column
    # name using [] operator
    columnSeriesObj = df_append[column]
    only_child = columnSeriesObj[:177].dropna()
    siblings = columnSeriesObj[177:].dropna()
    res = stats.mannwhitneyu(only_child, siblings)
    if res[1] <= 0.005:
        count += 1

    print('Column Name : ', column)
    print('Column Contents : ', columnSeriesObj.values)
    print('ratings for only child : ', only_child.values)
    print('ratings for siblings : ', siblings.values)
    print('results : ', res)
    print('Count for p value smaller than 0.005 : ', count)
```

```python
proportion = count/len(df_append)
proportion
```
