```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.stats.weightstats import ttest_ind
import pingouin as pg
```

```python
data_df = pd.read_csv('../movieReplicationSet.csv')
data_df
```

# Question5: Do people who are only children enjoy ‘The Lion King (1994)’ more than people with siblings?

First step: We seperate two groups for the only child and the child with siblings.

```python
only_child = data_df.loc[data_df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'] == 1, 'The Lion King (1994)'].values
```

```python
# drop the NaN values 
ii = np.isfinite(only_child)
only_child = only_child[ii]
only_child
```

```python
siblings = data_df.loc[data_df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'] == 0, 'The Lion King (1994)'].values
```

```python
# drop the NaN values 
ii = np.isfinite(siblings)
siblings = siblings[ii]
siblings
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")


sns.histplot(data=only_child, color="skyblue", label="The only child ratings", kde=True)
sns.histplot(data=siblings, color="red", label="The child with siblings ratings", kde=True)

plt.xlabel('Ratings')
plt.ylabel('Counts')
plt.title("Histogram of the Ratings for the only child and the child with siblings")
plt.legend()
plt.show()
```

Because of our assumptions and the analysis of how ratings should be analyzed, we will be using a Mann-Whitney U Test.

```python
stats.mannwhitneyu(only_child, siblings) 
```
