```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.stats.weightstats import ttest_ind
import pingouin as pg
```

```python
data_df = pd.read_csv('../movieReplicationSet.csv')
data_df
```


# Question7: Do people who like to watch movies socially enjoy ‘The Wolf of Wall Street (2013)’ more than those who prefer to watch them alone?

First step: We seperate two groups for the people watch alone and the people watch socially.

```python
enjoy_alone = data_df.loc[data_df['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)'] == 1, 'The Wolf of Wall Street (2013)'].values
enjoy_alone
```

```python
enjoy_social = data_df.loc[data_df['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)'] == 0, 'The Wolf of Wall Street (2013)'].values
enjoy_social
```

```python
# drop the NaN values 
ii = np.isfinite(enjoy_alone)
enjoy_alone = enjoy_alone[ii]
enjoy_alone
```

```python
# drop the NaN values 
ii = np.isfinite(enjoy_social)
enjoy_social = enjoy_social[ii]
enjoy_social
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")


sns.histplot(data=enjoy_alone, color="skyblue", label="The people who watch movies alone", kde=True)
sns.histplot(data=enjoy_social, color="red", label="The people who watch movies socially", kde=True)

plt.xlabel('Ratings')
plt.ylabel('Counts')
plt.title("Histogram of the Ratings for the The people who watch movies alone versus socially")
plt.legend()
plt.show()
```

```python
stats.mannwhitneyu(enjoy_alone, enjoy_social)
```
